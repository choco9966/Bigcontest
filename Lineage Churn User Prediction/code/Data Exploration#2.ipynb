```python
def generate_color():
    color = '#{:02x}{:02x}{:02x}'.format(*map(lambda x: random.randint(0, 255), range(3)))
    return color

def scatter_plot(data, x, y, color, showlegend1=False, name1 = "", xaxis1 = None, yaxis1 = None):
    trace = go.Scatter(
        x=data[x].values,
        y=data[y].values,
        name = name1,
        showlegend=showlegend1,
        marker=dict(
           color=color,
        ), opacity = 0.45, 
        xaxis = xaxis1,
        yaxis = yaxis1
    )
    return trace

def bar_plot(cnt_srs, color, name1 ="", xaxis1 = None, yaxis1 = None):
    trace = go.Bar(
            x = cnt_srs.index[::-1],
            y = cnt_srs.values[::-1],
            name = name1,
            textfont=dict(size=20),
            marker=dict(
                color=color,
            ), opacity = 0.45, 
            xaxis = xaxis1,
            yaxis = yaxis1
    )
    return trace
```

```python
# 서버의 종류를 분류해주는 함수 
def classification_server(dataframe):
    # 서버변수가 있는 dataframe이 들어오면, server가 특화, 일반, NON-PVP 
    # 셋 중 무엇인지 분류해주는 함수입니다. 
    PVP = ['bj', 'bk', 'bl']
    NON_PVP = ['be', 'bf', 'bg', 'bh', 'bi', 'bq', 'bs']
    GENERAL = ['aa','ab','ac','ad','ae','af','ag','ah','ai', 'aj', 'ak', 'al', 'am', 'an', 'ao', 'ap', 'aq', 'ar', 'as', 'at', 'au', 'av', 'aw', 'ax', 'ay', 'az', 'bb', 'bc', 'bd', 'bm', 'bn', 'bo', 'bp', 'br']
    
    # 만약 dataframe에 서버변수가 없다면 자기 자신을 출력해주도록 합니다. 
    if 'server' not in dataframe.columns.tolist():
        return dataframe 
    else :
        dataframe['server_class'] = dataframe['server'].apply(lambda x: 'PVP' if x in PVP else ('NON-PVP' if x in NON_PVP else 'GENERAL'))
        return dataframe
```

```python
# 서버의 종류를 분류해주는 함수 
def classification_level(dataframe):
    # 레벨변수가 있는 dataframe이 들어오면, 초보, 중수, 고수
    # 셋 중 무엇인지 분류해주는 함수입니다. 
    # 만약 dataframe에 레벨변수가 없다면 자기 자신을 출력해주도록 합니다. 
    if 'level' not in dataframe.columns.tolist():
        return dataframe
    elif 'acc_id' not in dataframe.columns.tolist():
        return dataframe
    else :
        dataframe_max_level = dataframe.groupby(['acc_id'])['level'].agg({'max'}).reset_index()
        dataframe_max_level['level_class'] = dataframe_max_level['max'].apply(lambda x: 'Hasu' if x <= 7 else ('Joongsu' if x <= 16 else 'Gosu'))
        dataframe_max_level = dataframe_max_level[['acc_id','level_class']]
        dataframe = dataframe.merge(dataframe_max_level,how='left',on=['acc_id'])
        return dataframe
```

```python
act_com_pay = act_com.merge(train_payment,how='left',on=['day','acc_id'])

# train_label의 amount_spent와 train_payment의 amount_spent가 겹치기때문에, label의 값을 바꿨습니다. 
train_label.columns = ['acc_id','survival_time','future_amount_spent']
act_com_pay = act_com_pay.merge(train_label,how='left',on=['acc_id'])

# train_label
act_com_pay['amount_spent'] = act_com_pay['amount_spent'].fillna(0)
act_com_pay['survival'] = act_com_pay['survival_time'].apply(lambda x: 1 if x == 64 else 0)
```

```python
act_com_pay_class_mean = act_com_pay.groupby(['day','level_class'])['amount_spent'].agg({'mean'}).reset_index()
act_com_pay_class_mean.columns = ['day','level_class','amount_spent_mean']
act_com_pay_class_mean = act_com_pay_class_mean.sort_values('level_class')

# Scatter plot by Groups
value = 'amount_spent_mean'

sns.lineplot(x = 'day', 
                y = value, 
                hue = 'level_class',  #different colors by group
                # style='species', # different shapes by group
                # s = 30, # marker size
                palette="Set1",                 
                data = act_com_pay_class_mean)

plt.show()
```

```python
# Scatter plot by Groups
act_com_pay = act_com_pay.sort_values('level_class')
value = 'survival_time'

sns.lineplot(x = 'day', 
                y = value, 
                hue = 'level_class',  #different colors by group
                # style='species', # different shapes by group
                # s = 30, # marker size
                palette="Set1",                 
                data = act_com_pay)

plt.show()
```
